## Scrapping Judul Berita

```python
# Scrapping semua judul berita pada halaman tersebut
for h_tag in soup.find_all(['h2','h3'], class_='elementor-post__title'):
    title = h_tag.a.text.strip()
    print(title)
```

Disini kita melakukan pengambilan seluruh judul berita yang ada pada halaman tersebut, setelah sebelumnya kita lakukan cek pada hasil soup halaman, kita mengatahui bahwa untuk judul sendiri itu berada pada tag html h2 dan h3 dengan class nya yaitu 'elementor-post__title' dan ada pada bagian tombol link 'a', sehingga untuk mengambil text dari tag tersebut kita menggunakan kode diatas dimana kita akan melakukan loop untuk setiap tag h2 dan h3 dengan class yang sesuai lalu kita ambil judul title dengan cara mengambil text.strip dari hasil loopingan sebelumnya

```python
from datetime import datetime

# Scrapping sesuai dengan tanggal hari ini
today_date = datetime.now().strftime('%d/%m/%Y')

for post in soup.find_all('div', class_='elementor-post__text'):
    title_element = post.find(['h2','h3'], class_='elementor-post__title')
    if title_element:
        title = title_element.a.text.strip()
    else:
        title = "Tidak ada judul"

    post_date_element = post.find('span', class_='elementor-post-date')
    if post_date_element:
        post_date = post_date_element.text.strip()
    else:
        post_date = "Tidak ada tanggal"

    if post_date == today_date:
        print(f'Tanggal: {post_date}')
        print(f'Judul: {title}\n')
```

Untuk mengambil judul berita dengan tanggal yang sama dengan tanggal hari ini yang perlu kita tambahkan dari kodingan sebelumnya adalah disini kita menambahkan variabel today_date dengan template day/month/year, lalu kita ambil semua judul dan tanggal dari html soup yang dimana kita lakukan looping pada tag div dengan class 'elementor-post__text'. lalu untuk pengambilan judul masih sama dengan sebelumnya dan untuk pengambilan tanggal disini kita ambil dari tag span dengan class 'elementor-post-date' dan kita lakukan percabangan jika tanggalnya sama dengan tanggal hari ini maka kita print judul berita beserta tanggalnya

# Nomor 2

```python
import requests
from bs4 import BeautifulSoup

# Loop through pages 1 to 209
for page_num in range(1, 210):
    url = f'https://unair.ac.id/category/featured/page/{page_num}/'
    response = requests.get(url)
    rawhtml = response.text
    soup = BeautifulSoup(rawhtml, 'html.parser')

    for h3_tag in soup.find_all('h3', class_='elementor-post__title'):
        title = h3_tag.a.text.strip()
        print(title)
```

Pada nomor ini kita akan melakukan scrapping data dari featured page berita, disini kita mengetahui bahwa ada 209 halaman featured page, untuk itu kita memerlukan perulangan dari halaman pertama sampai dengan halaman terakhir dari berita tersebut, kita mengetahui bahwa pada setiap halaman akan selalu mengganti angka dari https://unair.ac.id/category/featured/page/{page_num}/ sesuai dengan halaman ke berapanya, untuk itu kita bisa lakukan perulangan dari halama 1 sd halaman 209 untuk page_num nya. selanjutnya seperti biasanya untuk setiap halaman kita lakukan request pengambilan html, lalu parser dengan BeautifulSoap dan kita ambil data dari tag h3 dengan class 'elemtor-post__title' untuk menambil list judul dari setiap halaman

```python
# Ganti kodingan pada game_spider.py pada /content/myproject/myproject/spiders/game_spider.py

import scrapy

class GameSpiderSpider(scrapy.Spider):
    name = "game_spider"
    allowed_domains = ["bit.ly"]
    start_urls = ["https://bit.ly/scrapingtry"]

    def parse(self, response):
        # Use an XPath selector to find elements with the data-telemetry-meta attribute
        product_tiles = response.xpath('//a[@data-telemetry-meta]')

        for product_tile in product_tiles:
            data_telemetry_meta = product_tile.xpath('@data-telemetry-meta').get()
            self.log(data_telemetry_meta)  # You can log the data or process it further

            # You can parse the data as JSON to extract specific fields
            import json
            data_telemetry_meta_dict = json.loads(data_telemetry_meta)
            product_name = data_telemetry_meta_dict['name']
            product_price = data_telemetry_meta_dict['price']

            # Process or store the extracted data as needed
            yield {
                'product_name': product_name,
                'product_price': product_price,
            }
```
